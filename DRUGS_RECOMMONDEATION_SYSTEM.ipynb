```python
# prompt: write a code to take maximum value in poor , Excellent , Average and append that to new column Rating and make a label as poor,EXcellent and Average dont show me the values

# Create a new column 'Rating' with NaN values
df['Rating'] = np.nan

# Iterate through rows and assign the maximum value from 'Poor', 'Excellent', 'Average'
for index, row in df.iterrows():
    # Convert values to numeric before finding the max
    try:
        max_value = max([float(row['Poor']), float(row['Excellent']), float(row['Average'])])
    except ValueError:
        # Handle cases where conversion to float fails (e.g., non-numeric values)
        max_value = np.nan  # Or any other appropriate value

    df.loc[index, 'Rating'] = max_value

# Replace numerical values with labels
# Convert columns to numeric if needed
df['Poor'] = pd.to_numeric(df['Poor'], errors='coerce')
df['Excellent'] = pd.to_numeric(df['Excellent'], errors='coerce')
df['Average'] = pd.to_numeric(df['Average'], errors='coerce')

df['Rating'] = df['Rating'].replace(df['Poor'].max(), 'Poor').replace(df['Excellent'].max(), 'Excellent').replace(df['Average'].max(), 'Average')

# You can optionally drop the original 'Poor', 'Excellent', 'Average' columns if needed
# df = df.drop(['Poor', 'Excellent', 'Average'], axis=1)

print
```
